```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/titanic.csv")
data
```

```python
data.info()
```

```python
data.head()
```

```python
data.tail()
```

```python
data.columns
null_counts = data.isnull().sum()
null_counts
```

```python
null_counts.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Columns')
plt.ylabel('Number of Null Values')
plt.title('Null Values in Each Column')
plt.xticks(rotation=45)
plt.show()
```

```python
data['Age'].fillna(data['Age'].median(),inplace=True)
data
data['Survived']=data['Survived'].astype('category')
data['Pclass']=data['Pclass'].astype('category')
data['Sex']=data['Sex'].astype('category')
data['Age']=data['Age'].astype('int')
data['Embarked']=data['Embarked'].astype('category')
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=data)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', hue='Sex', data=data, palette='Set1')
plt.title('Survival Count by Sex')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Sex', loc='upper right')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=data, palette="cividis")
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x = 'Embarked', data=data, palette = 'Set1')
plt.title('Embarked Distribution')
plt.xlabel('Port of Embarkation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', data=data, palette = 'twilight')
plt.xlabel('Age')
plt.title('Box Plot: Distribution of Age')
plt.grid(True)
plt.show()
```

```python
data['FamilySize'] = data['SibSp'] + data['Parch'] + 1
plt.figure(figsize=(10, 6))
sns.countplot(x='FamilySize', hue='Survived', data=data, palette='Set1')
plt.title('Survival Count by Family Size')
plt.xlabel('Family Size')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=data, palette="Set2")
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sex', y='Age', hue='Survived', data=data)
plt.title('Age Distribution by Sex and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```
